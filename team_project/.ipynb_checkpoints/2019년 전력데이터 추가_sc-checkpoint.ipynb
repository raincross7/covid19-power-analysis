```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pymysql
import os
import warnings

plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 100
warnings.filterwarnings('ignore')
```

```python
# https://home.kepco.co.kr/kepco/KO/ntcob/list.do?boardCd=BRD_000097&menuCd=FN05030101
reader = pd.read_excel('datas/2019_12_sc.xlsx', sheet_name='21-1.산업분류별 행정구역별 판매전력량')
reader.to_csv('201912_power_data.csv')
```

```python
# csv 파일 다듬기
power = pd.read_csv('201912_power_data.csv')
power.columns = power.iloc[2]
power = power.drop([0,1,2])
```

```python
# DataFrame에 데이터 넣고 다듬기
city = [4,5,6,7,8,9,10,20,11,12,13,14,15,16,17,18,19]
code = [11,21,22,23,24,25,26,29,31,32,33,34,35,36,37,38,39]
year = [2019]*17
month = [12]*17
home = []
office = []
industry = []
for i in city:
    home.append(power.iloc[0,i])
    office.append(power.iloc[7,i])
    industry.append(power.iloc[40,i])

power_df = pd.DataFrame({'local_id': code, 'year': year, 'month': month, 'home': home, 'office': office, 'industry': industry})
power_df.home = power_df.home.astype(float).astype(int)
power_df.office = power_df.office.astype(float).astype(int)
power_df.industry = power_df.industry.astype(float).astype(int)
```

```python
power_df.to_csv('sale_by_201912_covid_sc.csv',index=False)
filePath = '201912_power_data.csv'
if os.path.exists(filePath):
    os.remove(filePath)
    print(f"{filePath} has been deleted")
else:
    print(f"{filePath} does not exist")
power_df
```
