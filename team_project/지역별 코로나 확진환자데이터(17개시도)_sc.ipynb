```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pymysql
from sqlalchemy import create_engine
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 100
```

```python
# 코로나19 양성자(표본) 감시현황
# https://ncov.kdca.go.kr/bdBoardListR.do?brdId=1&brdGubun=11
reader = pd.read_excel('datas/코로나바이러스감염증-19_확진환자_발생현황_230904_최종v2_sc.xlsx', sheet_name='시도별 발생(17개시도+검역)')
reader.to_csv('covid_rawdata.csv')
```

```python
# csv 파일 다듬기
covid = pd.read_csv('covid_rawdata.csv')
covid.columns = covid.iloc[3]
covid.drop([0,1,2,3,4], inplace=True)
covid.reset_index(inplace=True)
covid.일자 = pd.to_datetime(covid.일자)
covid.일자 = covid.일자.dt.strftime('%Y-%m')
covid
```

```python
# 2020-01 ~ 2023-08 까지
period_arr = []
for year in range(2020,2024):
    for month in range(1,13):
        if year==2023 and month==9:
            break
        period_arr.append(f'{year}-{month:02d}')
# 도시별, 년-월별 합계 0으로 초기화
cities_dic = {'서울': {}, '부산': {}, '대구': {}, '인천': {}, '광주': {}, '대전': {}, '울산': {}, '세종': {}, '경기': {},
              '강원': {}, '충북': {}, '충남': {}, '전북': {}, '전남': {}, '경북': {}, '경남': {}, '제주': {}}
for cityName in cities_dic:
    for period in period_arr:
        cities_dic[cityName][period] = 0
# 도시별, 년-월 별 합계
for i,row in covid.iterrows():
    for city in cities_dic:
        if row[city] != '-':
            cities_dic[city][row.일자] += int(row[city])
```

```python
# DB에 넣을 데이터로 가공
covid_df = pd.DataFrame({'local_id': [], '년-월': [], '확진자수': []})
cityCode_dic = {'서울': 11, '부산': 21, '대구': 22, '인천': 23, '광주': 24, '대전': 25, '울산': 26, '세종': 29, '경기': 31,
                '강원': 32, '충북': 33, '충남': 34, '전북': 35, '전남': 36, '경북': 37, '경남': 38, '제주': 39}

for cityName,cityCode in cityCode_dic.items():
    tmp_code = pd.DataFrame([cityCode]*len(cities_dic[cityName]))
    tmp_date = pd.DataFrame(cities_dic[cityName].keys())
    tmp_num = pd.DataFrame(cities_dic[cityName].values())
    tmp_cityData_df = pd.concat([tmp_code,tmp_date,tmp_num], axis=1)
    tmp_cityData_df.columns = ['local_id', '년-월', '확진자수']
    covid_df = pd.concat([covid_df,tmp_cityData_df], axis=0)
# 정수화
covid_df.local_id = covid_df.local_id.astype(int)
covid_df.확진자수 = covid_df.확진자수.astype(int)
```

```python
covid_df.to_csv('covid_data_sc.csv', index=False)
filePath = 'covid_rawdata.csv'
if os.path.exists(filePath):
    os.remove(filePath)
    print(f"{filePath} has been deleted")
else:
    print(f"{filePath} does not exist")
covid_df
```
